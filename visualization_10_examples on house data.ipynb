```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
%matplotlib inline
```

```python
data=pd.read_csv('data.csv')
data.head()
data.info()
```

1.Bar plot

```python
sns.barplot(data['sqft_living'], data['sqft_lot'], palette='viridis')
```

2.displot

```python
sns.displot(data['price'], bins=30, color='red')
```

3.joint plot

```python
sns.jointplot(data['bedrooms'], data['sqft_living'], kind='hex')
```

4.scatter plot

```python
sns.scatterplot(x='price', y='yr_built', data=data, color='red')
```

5.pairplot

```python
sns.pairplot(data)
```

```python
sns.pairplot(data[['sqft_basement','sqft_above','sqft_living']])
```

6.violin plot

```python
sns.violinplot(x='price',y='view',data=data, palette='plasma')
```

7.count plot

```python
sns.countplot(x ='floors', data = data, palette='plasma')
```

8.pointplot plot

```python
sns.pointplot(data['price'], data['view'], data['floors'])
```

9.Rug plot


```python
sns.rugplot(data['price'])
```

10.Box plot

```python
sns.boxplot(x='view', y='price', data=data)
```
