```python
class human:
  def __init__(this):
    this.name = "Adam smith"
    this.age = 1000
    this.currency = 'USD'
    this.hungerStatus = False
  def setNewProperties(this, name, age):
    this.name = name
    this.age = age
  def getHumanProperties(this):
    return this.name, this.age, this.currency, ("hungry" if this.hungerStatus==True else "Not Hungry")
```

```python
class ArithmeticProg:
  def __iter__(self):
    self.base = -3
    self.step = 2
    return self
  def __next__(self):
    self.base = self.base + self.step
    return self.base
```

```python
obj = ArithmeticProg()
i = iter(obj)
```

```python
for j in range(6):
  print(next(i))
```

```python
class Fibo:
  def __iter__(fibo):
    fibo.first = -1
    fibo.second = 1
    return fibo
  def __next__(fibo):
    fibo.third = fibo.first + fibo.second # 0
    fibo.first = fibo.second # 
    fibo.second = fibo.third

    return fibo.third
```

```python
i = iter(Fibo())
for j in range(10):
  print(next(i))
```
