Apply frequency distribution on each column.

Import important libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

call the dataset from scikitlearn dataset

```python
from sklearn.datasets import load_wine
```

```python
wine=load_wine()# just making an object
```

key of the dataset

```python
wine.keys()
```

```python
print(wine['DESCR']) #showing all the feature names
```

creating a dataframe

```python
data=pd.DataFrame(wine['data'],columns=wine['feature_names'])

data.head()
```

```python
data.tail()
```

```python
print(wine['target_names'])
```

add the class column in the table as dependent variable

```python
data['class']=wine['target']

data.head()
```

```python
data.tail()
```

Frequency

```python
sns.histplot(alcohol_freq, kde=True, alpha=0.7)
```
